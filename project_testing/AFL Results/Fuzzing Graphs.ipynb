```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# import data file
df = pd.read_csv("plot_data.txt")
df2 = pd.read_csv("old_data.txt")
```

```python
# compute paths that do not cause crash
df['paths_no_crash'] = df[' paths_total'] - df[' unique_crashes'] - df[' unique_hangs']
df['# unix_time'] = df['# unix_time'] - df['# unix_time'][0]
```

```python
df
```

```python
# Total paths against time
graph1 = df.plot(x='# unix_time', y=' paths_total', kind='line')
graph1.set_title('Total Paths')
graph1.set_xlabel('Unix Time')
graph1.set_ylabel('Total Paths')
graph1.get_legend().remove()
plt.show()
```

```python
# Unique crashes against time
graph2 = df.plot(x='# unix_time', y=' unique_crashes', kind='line')
graph2.set_title('Unique Crashes')
graph2.set_xlabel('Unix Time')
graph2.set_ylabel('Unique Crashes')
graph2.get_legend().remove()
plt.show()
```

```python
# Unique hangs against time
graph3 = df.plot(x='# unix_time', y=' unique_hangs', kind='line')
graph3.set_ylim(ymin=0)
graph3.set_title('Unique Hangs')
graph3.set_xlabel('Unix Time')
graph3.set_ylabel('Unique Hangs')
graph3.get_legend().remove()
plt.show()
```

```python
# Paths that do not cause crash against time
graph4 = df.plot(x='# unix_time', y='paths_no_crash', kind='line')
graph4.set_title('Paths that do not cause crash')
graph4.set_xlabel('Unix Time')
graph4.set_ylabel('Paths that do not cause crash')
graph4.get_legend().remove()
plt.show()
```
